```python
#Filtro los datos del csv población , excluyendo total pais
pobla_filtrado = poblacion[poblacion["provincia"] != "Total País"]
pobla_filtrado
```

```python
#Con la función merge combinamos los DataFrames, pobla_filtrado y hogares en un nuevo DataFrame llamado poblac_superf.
#La combinación se hará donde los valores de la columna provincia coincidan entre ambos DataFrames


poblac_superf = pd.merge(pobla_filtrado, hogares, left_on=["provincia"], right_on=["provincia"], how="left")
poblac_superf
```

```python
#Calculo la densidad, usando la superficie + la población total
poblac_superf["densidad_poblacion"]= poblac_superf["poblacion_total"]/poblac_superf["superficie_km2"]
poblac_superf
```

```python
#Comprobar que no haya datos nulos en mi dataset
poblac_superf.isnull().sum()
```

#Creación de columna ("fuera de rango")

crearemos una nueva columna "fuera_de_rango", que tomará del DataFrame "poblac_superf", los datos de la columna "densidad_de_poblacion", le resta la media(promedio) de todos los valores en la columna densidad_poblacion, utilizando np.mean y lo dividimos por la desviacion estándar calculada, aplicando np.std a "densidad_de_poblacion". Esto estandariza los valores, convirtiendolos e una medida llamada z-score. El z-score indica cuántas desviaciones estándar está un valor por encima o por debajo de la media

```python
#Estandarizando los valores con z-score
poblac_superf["fuera_de_rango"]=(poblac_superf["densidad_poblacion"] - np.mean(poblac_superf["densidad_poblacion"])) / np.std(poblac_superf["densidad_poblacion"])
poblac_superf
```

#Cálculo del Percentil

calculamos el percentil de la columna densidad de poblacion.

El percentil 99 es el valor por debajo del cual se encuentra en 99% de los datos. Esto significa que solo el 1% de los datos son mayores que este valor

```python
#Calculamos el percentil

p99 = np.percentile(poblac_superf["fuera_de_rango"], 99)
p99
```

```python
#Visualizar los valores (1%) que está fuera de rango
poblac_superf[poblac_superf["fuera_de_rango"] >= p99]
```

#Informe de variables originles

 partimos de la variable original "Hogares" para analizar la relación entre viviendas particulares totales y aquellas que se encuentran habitadas. A partir de este análisis, podemos afirmar que en Argentina, aproximadamente 1 de cada 6 viviendas no está habitada.

Este dato es del promedio nacional de viviendas no habitadas, que alcanza el 17,46%. Si invertimos este valor (100 / 17,46), obtenemos que por cada vivienda desocupada hay 5,7 habitadas, lo que nos indica  que 1 de cada 6 viviendas no está en uso.

podemos observar que:

El porcentaje mínimo de viviendas no habitadas en una provincia es de 12,05%

El máximo es de 23,95%

El promedio nacional es de 17,46%



```python
#Porcentaje de hogares no habitados
hogares["no_habitados"] = hogares["viviendas_particulares"] - hogares["viviendas_particulares_habitadas"]
hogares["porcentaje_de_no_habitados"] = (hogares["no_habitados"] / hogares["viviendas_particulares"]) * 100
hogares.head()
```

```python
#Media, minimo y maximo de hogares no habitados
print(f"Media: {hogares['porcentaje_de_no_habitados'].mean()}")
print(f"Minimo: {hogares['porcentaje_de_no_habitados'].min()}")
print(f"Maximo: {hogares['porcentaje_de_no_habitados'].max()}")
```

```python
#Porcentaje de viviendas no habitadas
porcentaje = hogares["porcentaje_de_no_habitados"].mean()
print("Porcentaje de viviendas no habitadas")
print(100/porcentaje)
```
